```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
data = sns.load_dataset("mpg")
data.head()
```

```python
cars_per_origin = pd.crosstab(data['name'], data['origin'], margins=True)
cars_per_origin
```

```python
cylinders_per_origin = pd.crosstab(data['origin'], data['cylinders'])
f, ax = plt.subplots(figsize=(12, 9))
cylinder_heatmap = sns.heatmap(cylinders_per_origin, cmap="YlGnBu", annot=True)
```

```python
# f, ax = plt.subplots(figsize=(12, 9))
heatmap = sns.heatmap(data.corr(numeric_only=True), vmin=-1, vmax=1, annot=True)
```
