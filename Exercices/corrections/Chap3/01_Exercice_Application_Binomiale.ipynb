```python
import random as r
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def bernouilli(p = .5):

    if r.random() > p:
        return 0

    return 1


def experience(n, p = .5, target = 1):
    count = 0

    for _ in range(n):
        if bernouilli(p) == target:
            count += 1

    return count


data = [ experience(10) for _ in range(1_000_000) ]
```

```python
# 3
count = 0
NB_SEARCH = 3

for _ in range(1_000_000):
    if experience(10) == NB_SEARCH:
        count += 1
```

```python
# Calcul empirique
print(count/1_000_000)

# 4

n, p, N = 10, 0.5, 1_000_000

# th√©orique avec numpy
((np.random.binomial(n, p, N) == NB_SEARCH).sum())/1_000_000
```
