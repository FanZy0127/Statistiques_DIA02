```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

p = { 'Prices' : ['210-230', '230-250', '250-270', '270-290'] , 'Effectifs' : [30, 60, 100, 20] }

prices = pd.DataFrame(p['Effectifs'], index=p['Prices'], columns=['Effectifs'])

prices.head()
```

```python
ax = prices.plot(figsize=(8,5), kind='bar', rot = 10)
ax.legend(loc="upper right", title="Prices", fontsize="large");
```
